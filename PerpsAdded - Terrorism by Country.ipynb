## Terrorism by Country

### a. Countries with the Most Terrorism Over the Years

```python
import pandas as pd

data = pd.read_csv('PerpsAdded.csv', encoding='ISO-8859-1', low_memory=False, sep=';')

country_counts = data['country_txt'].value_counts()

most_terrorism_country = country_counts.idxmax()
most_terrorism_count = country_counts.max()

print(f"The country with the most terrorism over the years is {most_terrorism_country} with {most_terrorism_count} events.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('PerpsAdded.csv', encoding='ISO-8859-1', low_memory=False, sep=';')

country_counts = data['country_txt'].value_counts()

top_10_countries = country_counts.head(10)

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top_10_countries.index, y=top_10_countries.values, palette="mako")

plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of Terrorism Events', fontsize=14)
plt.title('Top 10 Countries with Most Terrorism Events', fontsize=18)

plt.xticks(rotation=45, fontsize=12)

for i, v in enumerate(top_10_countries.values):
    ax.text(i, v + 25, str(v), horizontalalignment='center', fontsize=12)

    
plt.savefig('Top 10 Countries with Most Terrorism Events.png', dpi=300, bbox_inches='tight')
    
plt.tight_layout()
plt.show()

top_10_countries_table = top_10_countries.reset_index()
top_10_countries_table.columns = ['Country', 'Number of Terrorism Events']
print(top_10_countries_table)
```

### b. Top 10 Countries with the Most Terrorism in Recent Times (2012-2017)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('PerpsAdded.csv', encoding='ISO-8859-1', low_memory=False, sep=';')

data['date'] = pd.to_datetime(data['date'], format='%d.%m.%Y', errors='coerce')

recent_data = data[(data['date'].dt.year >= 2012) & (data['date'].dt.year <= 2017)]

recent_country_counts = recent_data['country_txt'].value_counts()

top_10_recent_countries = recent_country_counts.head(10)

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top_10_recent_countries.index, y=top_10_recent_countries.values, palette="mako")

plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of Terrorism Events', fontsize=14)
plt.title('Top 10 Countries with Most Terrorism Events (2012-2017)', fontsize=16)

plt.xticks(rotation=45)

for i, v in enumerate(top_10_recent_countries.values):
    ax.text(i, v + 50, str(v), horizontalalignment='center', fontsize=12)

plt.savefig('Top 10 Countries with Most Terrorism Events 2012-2017.png', dpi=300, bbox_inches='tight')

plt.show()

top_10_recent_countries_df = pd.DataFrame(top_10_recent_countries.reset_index())
top_10_recent_countries_df.columns = ['Country', 'Number of Terrorism Events']
print(top_10_recent_countries_df)
```

### c. Countries with Reduced Terrorism in Recent Times

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('PerpsAdded.csv', encoding='ISO-8859-1', low_memory=False, sep=';')

data['date'] = pd.to_datetime(data['date'], format='%d.%m.%Y', errors='coerce')

data_2007_2011 = data[(data['date'].dt.year >= 2007) & (data['date'].dt.year <= 2011)]
country_counts_2007_2011 = data_2007_2011['country_txt'].value_counts()

data_2012_2017 = data[(data['date'].dt.year >= 2012) & (data['date'].dt.year <= 2017)]
country_counts_2012_2017 = data_2012_2017['country_txt'].value_counts()

difference = country_counts_2007_2011 - country_counts_2012_2017
difference = difference.dropna().sort_values(ascending=False)

top_10_reduced_countries = difference.head(10)

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top_10_reduced_countries.index, y=top_10_reduced_countries.values, palette="mako")

plt.xlabel('Country', fontsize=14)
plt.ylabel('Reduction in Terrorism Events', fontsize=14)
plt.title('Top 10 Countries with Reduced Terrorism (2012-2017)', fontsize=16)

plt.xticks(rotation=45)

ax.set_ylim(0, max(top_10_reduced_countries.values) * 1.3)

for i, v in enumerate(top_10_reduced_countries.values):
    ax.text(i, v + 50, f'-{int(v)}', horizontalalignment='center', fontsize=12)

plt.savefig('Top 10 Countries with Reduced Terrorism 2012-2017.png', dpi=300, bbox_inches='tight')

plt.show()

top_10_reduced_countries_df = pd.DataFrame(top_10_reduced_countries).reset_index()
top_10_reduced_countries_df.columns = ['Country', 'Reduction in Terrorism Events']
print(top_10_reduced_countries_df)
```
